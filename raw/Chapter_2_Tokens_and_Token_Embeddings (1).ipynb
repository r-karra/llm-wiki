# Comparing Trained LLM Tokenizers


```python
from transformers import AutoModelForCausalLM, AutoTokenizer

colors_list = [
    '102;194;165', '252;141;98', '141;160;203',
    '231;138;195', '166;216;84', '255;217;47'
]

def show_tokens(sentence, tokenizer_name):
    tokenizer = AutoTokenizer.from_pretrained(tokenizer_name)
    token_ids = tokenizer(sentence).input_ids
    for idx, t in enumerate(token_ids):
        print(
            f'\x1b[0;30;48;2;{colors_list[idx % len(colors_list)]}m' +
            tokenizer.decode(t) +
            '\x1b[0m',
            end=' '
        )
```

```python
text = """
English and CAPITALIZATION
🎵 鸟
show_tokens False None elif == >= else: two tabs:"    " Three tabs: "       "
12.0*50=600
"""
```

# Recommending songs by embeddings

```python
import pandas as pd
from urllib import request

# Get the playlist dataset file
data = request.urlopen('https://storage.googleapis.com/maps-premium/dataset/yes_complete/train.txt')

# Parse the playlist dataset file. Skip the first two lines as
# they only contain metadata
lines = data.read().decode("utf-8").split('\n')[2:]

# Remove playlists with only one song
playlists = [s.rstrip().split() for s in lines if len(s.split()) > 1]

# Load song metadata
songs_file = request.urlopen('https://storage.googleapis.com/maps-premium/dataset/yes_complete/song_hash.txt')
songs_file = songs_file.read().decode("utf-8").split('\n')
songs = [s.rstrip().split('\t') for s in songs_file]
songs_df = pd.DataFrame(data=songs, columns = ['id', 'title', 'artist'])
songs_df = songs_df.set_index('id')
```

```python
print( 'Playlist #1:\n ', playlists[0], '\n')
print( 'Playlist #2:\n ', playlists[1])
```

```python
from gensim.models import Word2Vec

# Train our Word2Vec model
model = Word2Vec(
    playlists, vector_size=32, window=20, negative=50, min_count=1, workers=4
)
```

```python
song_id = 2172

# Ask the model for songs similar to song #2172
model.wv.most_similar(positive=str(song_id))
```

```python
print(songs_df.iloc[2172])
```

```python
import numpy as np

def print_recommendations(song_id):
    similar_songs = np.array(
        model.wv.most_similar(positive=str(song_id),topn=5)
    )[:,0]
    return  songs_df.iloc[similar_songs]

# Extract recommendations
print_recommendations(2172)
```

```python
print_recommendations(2172)
```

```python
print_recommendations(842)
```

## Navigating the Hugging Face Ecosystem

Hugging Face is a central hub for machine learning, especially for natural language processing. It provides:

*   **🤗 Transformers library**: The library you're already using to load models and tokenizers (like `AutoModelForCausalLM` and `AutoTokenizer`). It supports PyTorch, TensorFlow, and JAX.
*   **🤗 Datasets library**: A fast and efficient library for loading and processing datasets.
*   **🤗 Accelerate library**: Simplifies distributed training for various hardware.
*   **Hugging Face Hub**: A platform for sharing models, datasets, and demos (Spaces).

You've already engaged with the Hub by downloading `microsoft/Phi-3-mini-4k-instruct` and are using the `transformers` library for core model operations. Let's explore a more streamlined way to perform text generation using the `pipeline` feature from the `transformers` library.

```python
from transformers import pipeline, AutoModelForCausalLM, AutoTokenizer
import torch
import os

# Redefine model_local_path to ensure it's available
drive_model_path = '/content/drive/MyDrive/colab_data/models'
model_name = "microsoft/Phi-3-mini-4k-instruct"
model_local_path = os.path.join(drive_model_path, model_name.replace("/", "--"))

# Load model and tokenizer from the local path in Google Drive
model = AutoModelForCausalLM.from_pretrained(
    model_local_path,
    device_map="cuda",
    torch_dtype="auto",
    trust_remote_code=False,
)
tokenizer = AutoTokenizer.from_pretrained(model_local_path)

# Create a text generation pipeline using the already loaded model and tokenizer
generator = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    device=0 # Use GPU (device_map='cuda' would also work if model wasn't already moved)
)

# Define a prompt
prompt = "Tell me a short story about a brave knight and a dragon.<|assistant|>"

# Generate text using the pipeline
response = generator(prompt, max_new_tokens=50, num_return_sequences=1)

# Print the generated text
print(response[0]['generated_text'])
```

This `pipeline` abstraction simplifies the process of text generation, handling tokenization, model inference, and decoding all in one go. It's a very common and convenient way to interact with models on Hugging Face.
